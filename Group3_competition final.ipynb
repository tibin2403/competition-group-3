Time series plots of sensor data shows that the sensor_15 does not contain any data.
Considering the data pattern between some sensors, they are very similar to each other.Sensor_50 doesn't contain any data after 140000 onwards.There is a period of missing data for sensor_51

```python
data=data.drop(['sensor_15','sensor_50','sensor_51'],axis=1)
data.head()
```

```python
data.isna().sum()
```

```python
#Filling the null values
```

```python
df=data.iloc[:,1:50].fillna(method = 'ffill')
df.head()
```

```python
df.isna().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Checking the outliers
```

```python
numcols = data.dtypes[(data.dtypes=='int64') | (data.dtypes=='float64')].index
numcols
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

```python
for x in numcols:
    print(x)
    descr = data[x].describe()
    IQR = descr['75%'] - descr['25%']
    upper_limit =  descr['75%'] + 1.5 * IQR
    lower_limit =  descr['25%'] - 1.5 * IQR  
    #outlier replacement
    data[x][data[x]>upper_limit] = upper_limit
    data[x][data[x]<lower_limit] = lower_limit
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

```python
plt.title('machine_status',fontsize=19)
sns.countplot(data.machine_status)
plt.show()
```

```python
#Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
data['machine_status']=label_en.fit_transform(data['machine_status'])
data.head()
```

```python
#Splitting the data
```

```python
X=df
X.head()
```

```python
y=data['machine_status']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#Modelling
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(X_train,y_train)
y_pred=logit_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
y_pred=dt_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#Fine tuning
```

```python
rft=RandomForestClassifier(n_estimators=500,max_depth=10,random_state=42,criterion='entropy')
rft.fit(X_train,y_train)
y_pred=rft.predict(X_test)
accuracy_score(y_test,y_pred)
```
